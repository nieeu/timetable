```python
df_sub = df[['course_id', 'class_id','credits', 'period', 'day', 'start_slot', 'class_code', 'day_slot']]
df_sub.head()
```

```python
df_per1 = df_sub[
    df['period']==1
]
df_per1.head()
```

```python
df_per1['class_code_full'] = df_per1['course_id'] +'.' +df_per1['class_code'].astype(str)
df_per1
```

```python
course_list = ['KDO402', 'TMA404', 'TAN432', 'TMA311', 'TMA408']

df_class = df_per1[
    df_per1['course_id'].isin(course_list)
]
df_class['day'][df_class['day_slot']=='Thứ 2(01-03)'] = '2Ca1'
df_class['day'][df_class['day_slot']=='Thứ 2(07-09)'] = '2Ca3'
df_class['day'][df_class['day_slot']=='Thứ 2(04-06)'] = '2Ca2'
df_class
```

```python
set1 = df_class['class_code_full'][df_class['course_id']=='KDO402'].unique().tolist()
set2 = df_class['class_code_full'][df_class['course_id']=='TMA404'].unique().tolist()
set3 = df_class['class_code_full'][df_class['course_id']=='TAN432'].unique().tolist()
set4 = df_class['class_code_full'][df_class['course_id']=='TMA311'].unique().tolist()
set5 = df_class['class_code_full'][df_class['course_id']=='TMA408'].unique().tolist()

all_list = []

import itertools
list_set = [set1, set2, set3, set4, set5]
for element in itertools.product(*list_set):
  all_list.append(element)

all_list
```

```python
class_code_list = df_class['class_code_full'].unique().tolist()
```

```python
df_week['class_code'] = class_code_list
```

```python
df_week[[slots]] = 0
```

```python
for a in class_code_list:
    ca_list = df_class['day'][
        df_class['class_code_full'] == a
    ].tolist()

    for x in ca_list:
        df_week[x][
            (df_week['class_code'] == a)
        ] += 1

df_week
```
